Recolección de los datos de la página

```python
from ucimlrepo import fetch_ucirepo 
#ucimlrepo importar  conjuntos de datos del Repositorio de Aprendizaje de Máquinas Irvine de UC en scripts y cuadernos. 
#fetch_ucirepo Carga los datos del Repositorio ML de la UCI, incluye:  fotogramas de datos y la información de metadatos.
#metadato informacion que describe los datos

# fetch dataset
cdc_diabetes_health_indicators = fetch_ucirepo(id=891)  #id del Dataset para el repositorio ML del UCI
#DataSet conjunto de datos en donde se almacena la información( tablas, archivos, hojas de cálculo, coleccion de imagenes)

# data (as pandas dataframes)
X = cdc_diabetes_health_indicators.data.features #conjunto de características 
y = cdc_diabetes_health_indicators.data.targets #target, valor objetivo

# metadata (infomración de los datos )
print(cdc_diabetes_health_indicators.metadata)

# variable information
print(cdc_diabetes_health_indicators.variables)
#demographic describen su estructura en términos e diferentes atributos personales
```

Visualización de las variables

```python
print("Variables características")
print(X[:5])
print("Variables objetivo")
print(y[:5])
```

Conversión a un DataFrame

```python
import pandas as pd #importacion de pandas
#conversion a dataFrame
X = pd.DataFrame(X)
y = pd.DataFrame(y)
data_Frame = pd.concat([X, y], axis=1) 
data_Frame.head()
```

```python
#Valores únicos
for column in data_Frame.columns:
    print(f'Columna: {column}, Valores únicos: {data_Frame[column].unique()}')
```

### Clasificador A: 
El proyecto 1 el cual fue un proyecto relacionado a los árboles de decisión, las características elegidas por el clasificador fueron: GenHlth, HighBP. 
En este caso únicamente se elegiran dichas características para poder realizar el clasificador. 


```python
caracteristicas = ['GenHlth','HighBP']
df_caracteristicas = data_Frame[caracteristicas]

df_caracteristicas.head()
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.naive_bayes import MultinomialNB

# Configurar K-Fold Cross Validation con K=5
kf_a = KFold(n_splits=5, shuffle=True, random_state=42)

# Inicializar el modelo Bayes ingenuo multinomial
model_a = MultinomialNB()
```

```python
# Realizar validación cruzada y calcular la precisión promedio
X_a = df_caracteristicas
target = data_Frame['Diabetes_binary']
scores = cross_val_score(model_a, X_a, target, cv=kf_a, scoring='accuracy')

# Imprimir resultados globales
print(f"Precisión promedio (5-fold CV): {scores.mean():.4f}")
print(f"Desviación estándar: {scores.std():.4f}")
```

### Clasificador B: 
Incluye sólo las características: HighBP,   HighChol,   HeartDiseaseorAttack,  HvyAlcoholConsump,     GenHlth,    DiffWalk. Además realiza una evaluación  K-fold cross validation con  K=5

Primero se seleccionan las características con las que se trabajará

```python
# Seleccionar las características relevantes para el Clasificador B
features_B = ["HighBP", "HighChol", "HeartDiseaseorAttack", "HvyAlcoholConsump", "GenHlth", "DiffWalk"]
X_nueva = data_Frame[features_B]  # Datos de entrada (características seleccionadas)
y_nueva = data_Frame["Diabetes_binary"]  # Etiqueta objetivo (Diabetes_binary)
```

```python
# Verificar la selección de datos
nueva_data_Frame= pd.concat([X_nueva, y_nueva], axis=1)
nueva_data_Frame.head()
```

Si las características seleccionadas sean categóricas, se convierten en numéricas. Dato que los datos ya contienen valores numéricos no será necesario realizarlo, sin embargo se hace la práctica de la tranformación de los datos. 

```python
from sklearn.preprocessing import LabelEncoder 

# Convertir las características seleccionadas a valores numéricos si no lo están
transfomracion_informacion = {}
for column in X_nueva.columns:
    le = LabelEncoder()  # Inicializar un codificador
    X_nueva[column] = le.fit_transform(X_nueva[column])   # Transformar los valores categóricos a numéricos
    transfomracion_informacion[column] = le  # Guardar el codificador para posibles usos futuros

# Verificar el preprocesamiento
X_nueva.head()
```

Ahora se realiza la configuración para la validación cruzada

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.naive_bayes import MultinomialNB

# Configurar K-Fold Cross Validation con K=5
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Inicializar el modelo Bayes ingenuo multinomial
model = MultinomialNB()
```

Evaluación del modelo

```python
# Realizar validación cruzada y calcular la precisión promedio
scores = cross_val_score(model, X_nueva, y_nueva, cv=kf, scoring='accuracy')

# Imprimir resultados globales
print(f"Precisión promedio (5-fold CV): {scores.mean():.4f}")
print(f"Desviación estándar: {scores.std():.4f}")
```
